== EDA ==


```python
EV.describe().plot(kind='bar')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='County',data=EV)
plt.ylabel('Frequency')
plt.title('the country wise list')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(EV.corr(numeric_only=True),annot=True,cmap='plasma')
plt.title('The heatmap data info')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Make',y='Base MSRP',data=EV)
plt.title('The make wise base')
plt.legend()
plt.show()
```

```python
category_col = EV.select_dtypes(include='object').columns

plt.figure(figsize=(10,6))
for col in category_col:
    sns.countplot(data=EV,x=col)
    plt.title(f'The distribution of{col}')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
plt.figure(figsize=(10,6))
top_make = EV.groupby('Make')['Electric Range'].sum()
top_make.plot(kind='line',color='y')
plt.title('The top make vehicals')
plt.show()
```

```python
numerical_df = EV.select_dtypes(include=np.number).columns

plt.figure(figsize=(10,6))
for col in numerical_df:
    sns.histplot(x=col,data=EV,kde=True,color='salmon')
    plt.title(f'The distribution of {col}')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.pairplot(data=EV,markers='*',hue='Make')
plt.title('The pairplot data info')
plt.show()
```
